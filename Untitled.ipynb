## Q1. In which year was published most of films and serials

```python
#Group by 
subdf = df.groupby(['release_year']).count().drop(columns=['show_id',
                   'type',
                   'title',
                   'director',
                   'description\t\t',
                   'country',
                   'date_added',
                   'rating',
                   'duration',
                   'listed_in'
                  ]).reset_index()
subdf 
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax = sns.barplot(data=subdf, x='release_year',y='cast',ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90);
```

## Q2. In what country was made majority of movies

```python
subdf = df[['country','show_id']]
subdf
```

```python
subdf = subdf.set_index(['show_id']).apply(lambda x: x.str.split(',').explode()).reset_index().groupby('country').count().reset_index().sort_values(by='show_id', ascending=False)
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax = sns.barplot(data=subdf.head(20), x='country',y='show_id',ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90);
```

## Q3. Get movie with biggest duration

```python
subdf1 = df[df['listed_in']=='Movies'].sort_values(by='duration', ascending=False)['duration'].str.split(' ').explode()
subdf1[subdf1!='min']
subdf2 = df[df['listed_in']=='Movies']
subdf2['duration'] = subdf1[subdf1!='min']
subdf2
```

```python
fig, ax1 = plt.subplots(figsize=(20,10))
ax1 = sns.barplot(data=subdf2.sort_values(by=['duration']), x='title',y='duration',ax=ax1,color='tab:cyan')
ax1.set_xticklabels(ax.get_xticklabels(), rotation=90);
ax2 = ax1.twinx()
ax2 = sns.lineplot(x='title',y='release_year',data=subdf2, sort=False, color='0')
plt.show()
```

## Q4. Get count of tv shows by number of seasons 

```python
subdf1 = df[df['type']=='TV Show'].sort_values(by='duration', ascending=False)['duration'].str.split(' ').explode()
subdf1 = subdf1[subdf1!='Seasons']
subdf1 = subdf1[subdf1!='Season']
subdf2 = df[df['type']=='TV Show']
subdf2['duration'] = subdf1.astype(int)
subdf2
```

```python
subdf2.sort_values(by=['duration'])
```

```python
subdf2.groupby('duration').count().reset_index()
```

```python
fig, ax1 = plt.subplots(figsize=(20,10))
ax1 = sns.barplot(data=subdf2.groupby('duration').count().reset_index(), x='duration', y='show_id',ax=ax1)
```

## Q5. In what month was filmed majority of movies and serials

```python
list_data = [df.date_added.str.split(' ')[index][0]
             for index in df.index]
sdf = pd.DataFrame(list_data)
subdf = df
subdf['date_added'] = sdf
subdf
```

```python
subdf = subdf.dropna()
subdf = subdf[subdf['date_added']!='']
subdf
```

```python
subdf.groupby(['type','date_added']).count().drop(columns=['title','director','cast','country','release_year','rating','duration','listed_in','description\t\t']).reset_index()
```

```python
fig, ax1 = plt.subplots(figsize=(20,10))
ax1 = sns.barplot(data=subdf[subdf['type']=='Movie'], x='date_added', y='show_id', ax=ax1, ci=None)
```

```python
fig, ax1 = plt.subplots(figsize=(20,10))
ax1 = sns.barplot(data=subdf[subdf['type']=='TV Show'], x='date_added', y='show_id', ax=ax1, ci=None)
```
